```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
Weeks = ['pre-season', 'week-2', 'week-3', 'week-4', 'week-5', 'week-6', 'week-7', 'week-8', 'week-9', 'week-10', 'week-11', 'week-12', 'week-13', 'week-14', 'week-15', 'week-16', 'final-rankings']
url = "https://collegepolltracker.com/football/grid/2018/"
Dictionaries = []
voters = set()
teams = set()
for week in Weeks:
    new_url = url + week
    response = requests.get(new_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
    else:
        print("bad")
        continue
    parent_element = soup.find('div', id='gridBallots')
    Dict = {}
    if parent_element:
        Rows = parent_element.find_all('div', class_='gridRow')
        for row in Rows:
            voter_name = (row.find('div', class_='gridPollster').find('a'))['href']
            Team_rows = (row.find_all('div', class_='gridTeam'))
            for div in Team_rows:
                correct_team_name = div.get('class')[1][3:]
                if correct_team_name != 'blank': 
                    teams.add(correct_team_name)
            correct_voter_name = voter_name.split("/")[3]
            voters.add(correct_voter_name)

alphabetic_order_voters = list(voters)
alphabetic_order_voters.sort()

alphabetic_order_teams = list(teams)
alphabetic_order_teams.sort()

print("Voters: ", alphabetic_order_voters)
print("Teams: ", alphabetic_order_teams)
```

```python
def write_to_file(file_path, list_of_data):
    with open(file_path, 'w') as file:
        for data in list_of_data:
            file.write(data + "\n")
```
