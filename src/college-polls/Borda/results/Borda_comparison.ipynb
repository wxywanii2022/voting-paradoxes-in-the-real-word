```python
import pandas as pd
import os

def ranking_comp(year, week, voting_method):
    #Reads in borda count data corresponding with year, week, and the provided voting method if it exists
    try:
        top25_df = pd.read_csv(f"./borda_top25/season_{year}/{year}_week{week}_top25.csv")
        vote_method = pd.read_csv(f"./borda_{voting_method}/season_{year}/{year}_week{week}_{voting_method}.csv")
    except:
        print(f'Data for this combination of week and year (year: {year}, week: {week}) was not available in the local ballot_data_by_season_and_week directory')
        return

    #Captures team rankings of standard borda count and the provided alternate borda count method
    top25_teams = top25_df["Teams"].tolist()
    new_method_teams = vote_method["Teams"].tolist()

    results = []

    #Checks for the highest differing ranking by comparing list values per iteration
    for idx in range(min(len(top25_teams), len(new_method_teams))):
        if top25_teams[idx] != new_method_teams[idx]:
            results.append({
                'Season': year,
                'Week': week,
                'Highest Diff Ranking': idx + 1
            })
            break

    return pd.DataFrame(results)


def csv_maker(voting_method):
    all_results_df = pd.DataFrame()

    #Iterate through all combinations of weeks and years in order to craft borda count comparison csv for the provided voting method
    for year in range(2014, 2025):
        for week in range(1, 18):
            result_df = ranking_comp(str(year), str(week), voting_method)
            all_results_df = pd.concat([all_results_df, result_df], ignore_index=True)

    all_results_df.sort_values("Highest Diff Ranking", axis=0, inplace=True, ascending=True)
    os.makedirs(f"../borda_comparator/highest_rank_diff", exist_ok=True)
    all_results_df.to_csv(f"../borda_comparator/highest_rank_diff/{voting_method}_borda_comparison.csv", index=False)

voting_methods = ["top10", "top5", "top3", "top1", "Dowdall"]

for method in voting_methods:
    csv_maker(method)
```
